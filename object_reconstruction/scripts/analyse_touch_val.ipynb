```python
import numpy as np
import object_reconstruction.data.checkpoints as checkpoints
import os
import plotly.graph_objects as go
```

```python
filename = 'val_dict_30_08_1304.npy'

file_path = os.path.join(os.path.dirname(checkpoints.__file__), 'touch_val_debug', filename)
val_dict = np.load(file_path, allow_pickle=True).item()
```

Plot predicted pointcloud and ground truth pointcloud of the first element in the dictionary:

```python
pred_points = val_dict['pred_points'][5]
pred_verts = val_dict['pred_verts'][5]
pointcloud_gt = val_dict['pointcloud_gt'][5]
if len(pred_points.shape)>2:
    pred_points = pred_points.reshape(pred_points.shape[0]*pred_points.shape[1], pred_points.shape[2])
    pred_verts = pred_verts.reshape(pred_verts.shape[0]*pred_verts.shape[1], pred_verts.shape[2])
    pointcloud_gt = pointcloud_gt.reshape(pointcloud_gt.shape[0]*pointcloud_gt.shape[1], pointcloud_gt.shape[2])
print(pred_points.shape)
fig = go.Figure(
    [
        go.Scatter3d(x=pred_points[:, 0], y=pred_points[:, 1],z=pred_points[:, 2], mode='markers',marker=dict(size=1)),
        go.Scatter3d(x=pointcloud_gt[:, 0], y=pointcloud_gt[:, 1],z=pointcloud_gt[:, 2], mode='markers',marker=dict(size=1)),
        #go.Scatter3d(x=pred_verts[:, 0], y=pred_verts[:, 1],z=pred_verts[:, 2], mode='markers',marker=dict(size=5))
    ]
)
fig.show()
```
