```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sample_hemisphere(r):
    """
    Uniform sampling on a hemisphere.
    Parameter:
        - r: radius
    """
    theta = 2 * np.pi * np.random.uniform()
    phi = np.arccos(1 - 2 * np.random.uniform())
    x = r * np.sin(phi) * np.cos(theta)
    y = r * np.sin(phi) * np.sin(theta)
    z = np.absolute(r * np.cos(phi))
    return x, y, z
```

```python
r = 0.5
x_list = []
y_list = []
z_list = []

for i in range(0, 500):
    x, y, z = sample_hemisphere(r)
    x_list.append(x)
    y_list.append(y)
    z_list.append(z)
# Attaching 3D axis to the figure
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(projection="3d")
scatters = ax.scatter(x_list, y_list, z_list, s=2)
ax.set_xlim(-0.5,0.5)
ax.set_ylim(-0.5,0.5)
ax.set_zlim(-0.5,0.5)
plt.show()
```
