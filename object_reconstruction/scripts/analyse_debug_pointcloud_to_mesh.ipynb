# Analyse debug pointcloud to mesh

The purpose of this notebook is to plot the filtered pointcloud and the resulting mesh obtained in `extract_touch_charts`. 
In `sample_utils.py` $\rightarrow$ `spherical_sampling()`, I store optionally a dictionary called `debug_pointcloud_to_mesh.npy`. 
This dictionary is structred as follows:
```
- debug_pointcloud_to_mesh
    |- {obj_index}
        |- list [pointcloud, mesh]
```
In the list: `pointcloud` is the filtered pointcloud with variable size, `mesh` should contain only 25 vertices

````python
import numpy as np
import os
import object_reconstruction.data.checkpoints as checkpoints
import plotly.graph_objects as go
````

````python
dictionary_path = os.path.join(os.path.dirname(checkpoints.__file__), 'debug_pointcloud_to_mesh', 'debug_pointcloud_to_mesh.npy') 
dictionary = np.load(dictionary_path, allow_pickle=True).item()
dictionary.keys()
````

````python

keys = list(dictionary['102408'].keys())

for key in keys:
    pointcloud = dictionary['102408'][key][0] 
    mesh = dictionary['102408'][key][1][0]
    print(mesh.shape)

    fig = go.Figure(
        [   go.Scatter3d(x=pointcloud[:, 0], y=pointcloud[:, 1], z=pointcloud[:, 2], mode='markers',marker=dict(size=1)),
            go.Scatter3d(x=mesh[:, 0], y=mesh[:, 1], z=mesh[:, 2], mode='markers', marker=dict(size=1, color='orange')),
        ]
    )
    fig.show()
     
````
